```python
import cv2
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop
import numpy as np
```

```python
path = "images/image4.jpg"

img, image = cv2.imread(path),cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv_pixel = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
blur = cv2.medianBlur(image, 9)
gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
img_canny = cv2.Canny(otsu, 10, 10)

cnts = cv2.findContours(img_canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]


min_area = 15
dots = []
for c in cnts:
    area = cv2.contourArea(c)
    if area > min_area:
        cv2.drawContours(img, [c], -1, (100, 255, 22), 2)
        dots.append(c)

print("White Dots count is:",len(dots))

pixel = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# create figure 
fig = plt.figure(figsize=(15, 15)) 

val = [(image,1),(blur,2),(gray,3),(otsu,4),(img_canny,5),(img,6),(pixel,7)]


for v,i in val:
    fig.add_subplot(3, 3, i) 
    # showing image 
    plt.imshow(v) 
    plt.axis('off') 
    plt.title(i) 
```
